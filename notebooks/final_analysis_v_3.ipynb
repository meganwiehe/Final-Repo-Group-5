```python
inspected_restaurants = hd.groupby('dba_name').filter(lambda x: len(x) > 1)
```

```python
inspected_restaurants['Upgrade/Downgrade'] = np.where(inspected_restaurants['results'].shift(-1) != inspected_restaurants['results'], 'Yes', 'No')
```

```python
inspected_restaurants['Pass/Fail'] = np.where(inspected_restaurants['results'] == 'Pass', 'Pass', 'Fail')
```

```python
inspected_restaurants['Result Numerical'] = (inspected_restaurants['results'] == 'Pass').astype(int)
```

```python
restaurant_inspection_counts = inspected_restaurants.groupby('dba_name').size().reset_index(name='Inspection Count')
```

```python
inspection_scores = inspected_restaurants.groupby('dba_name')['Result Numerical'].agg(['mean', 'median']).reset_index()
```

```python
restaurant_data = pd.merge(restaurant_inspection_counts, inspection_scores, on='dba_name', how='left')
```

```python
restaurant_data['Inspection Summary'] = restaurant_data['dba_name'] + ' - Inspected ' + restaurant_data['Inspection Count'].astype(str) + ' times'
```

```python
print(f"Total number of inspected restaurants: {len(inspected_restaurants)}")
```

```python
upgrade_downgrade_counts = inspected_restaurants['Upgrade/Downgrade'].value_counts()
```

```python
pivot_table = inspected_restaurants.pivot_table(index='dba_name', columns='results', aggfunc='size', fill_value=0)
print("Pivoted table:")
print(pivot_table)
```

```python
upgrade_downgrade_counts = inspected_restaurants.groupby(['dba_name', 'Upgrade/Downgrade']).size().unstack(fill_value=0).reset_index()
upgrade_downgrade_counts = upgrade_downgrade_counts.rename(columns={'No': 'Upgraded', 'Yes': 'Downgraded'})
```

```python
pass_fail_counts = inspected_restaurants.groupby(['dba_name', 'Pass/Fail']).size().unstack(fill_value=0).reset_index()
```

```python
merged_data = restaurant_inspection_counts.merge(upgrade_downgrade_counts, on='dba_name', how='left')
merged_data = merged_data.merge(pass_fail_counts, on='dba_name', how='left')
```

```python
merged_data.to_csv('restaurant_inspection_summary.csv')
```
